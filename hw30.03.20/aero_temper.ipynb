```python
data=[]
from itertools import groupby
with open('temper.stat.txt','r') as file:
    reader = file.readlines()
    for line in reader:
        data.append(float(line.strip()))
print(f'макс.:{max(data)}\nмин.: {min(data)}\nср. знач.:{sum(data)/len(data)}\nуникальных температур: {len(list(groupby(data)))} из {len(data)}')
```

```python
x_s=list(range(1,len(data)+1))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(x_s, data)
plt.show()
```
